```python
from PIL import Image
from logic_operation import operate_logical_not, operate_logical_operation
from util import display_image

# Load the RGB images
triangle_image = Image.open("triangle.jpg")
circle_image = Image.open("circle.jpg")


and_image = operate_logical_operation(triangle_image, circle_image, "AND")
or_image = operate_logical_operation(triangle_image, circle_image, "OR")
xor_image = operate_logical_operation(triangle_image, circle_image, "XOR")
not_image = operate_logical_not(triangle_image)

display_image([and_image, or_image, xor_image, not_image], 4, 1)
```
