```python
import cv2
import numpy as np
from util import plot_images

# Membaca gambar
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

kernel = np.array([[-1, -1, -1],
                  [-1,  8, -1],
                  [-1, -1, -1]])

# Mengaplikasikan kernel ke gambar
edge_detected = cv2.filter2D(image, -1, kernel)

result = [image]
result_name = ['Gambar Asli']

for threshold_value in range(7):  # 0 sampai 7
    threshold_value /= 7
    _, thresholded_image = cv2.threshold(edge_detected, threshold_value * 255, 255, cv2.THRESH_BINARY)
    result.append(thresholded_image)
    result_name.append(f'Treshold {threshold_value:.2f}')

plot_images(result, result_name,  4, 2)
```
