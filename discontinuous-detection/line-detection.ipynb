```python
import cv2
import numpy as np
from util import plot_images
# Membaca gambar
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Kernel untuk deteksi garis horisontal
kernel_horizontal = np.array([[-1, -1, -1],
                              [2, 2, 2],
                              [-1, -1, -1]])

# Kernel untuk deteksi garis vertikal
kernel_vertical = np.array([[-1, 2, -1],
                            [-1, 2, -1],
                            [-1, 2, -1]])

# Kernel untuk deteksi garis 45 derajat
kernel_45_degrees = np.array([[-1, -1, 2],
                              [-1, 2, -1],
                              [2, -1, -1]])

# Kernel untuk deteksi garis 135 derajat
kernel_135_degrees = np.array([[2, -1, -1],
                                [-1, 2, -1],
                                [-1, -1, 2]])


# Mengaplikasikan kernel ke gambar
horizontal_lines = cv2.filter2D(image, -1, kernel_horizontal)
vertical_lines = cv2.filter2D(image, -1, kernel_vertical)
lines_45_degrees = cv2.filter2D(image, -1, kernel_45_degrees)
lines_135_degrees = cv2.filter2D(image, -1, kernel_135_degrees)

# Temukan nilai maksimum dari setiap hasil konvolusi
max_horizontal = np.max(horizontal_lines)
max_vertical = np.max(vertical_lines)
max_45_degrees = np.max(lines_45_degrees)
max_135_degrees = np.max(lines_135_degrees)

# Tentukan arah garis yang mendominasi
dominant_direction = np.argmax([max_horizontal, max_vertical, max_45_degrees, max_135_degrees])

# Menampilkan hasil deteksi garis
result = [image, horizontal_lines, vertical_lines, lines_45_degrees, lines_135_degrees]
result_name = ['Original', 'Horizontal Lines', 'Vertical Lines', '45 Degrees Lines', '135 Degrees Lines']


plot_images(result, result_name, 5, 1)

directions = ["Horizontal", "Vertical", "45 Degrees", "135 Degrees"]
```
