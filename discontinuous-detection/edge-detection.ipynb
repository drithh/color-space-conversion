```python
import cv2
import numpy as np
from util import plot_images, normalize
%load_ext autoreload
%autoreload 2


image = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Sobel operator using NumPy
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

# Convolve the image with Sobel filters
sobel_x_result = cv2.filter2D(image, cv2.CV_64F, sobel_x)
sobel_y_result = cv2.filter2D(image, cv2.CV_64F, sobel_y)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(sobel_x_result**2 + sobel_y_result**2)
# show vertical edges normalize

result = [image, normalize(sobel_x_result), normalize(sobel_y_result), normalize(gradient_magnitude)]
result_name = ["Original Image", "Sobel X", "Sobel Y", "Gradient Magnitude"]

plot_images(result, result_name)
```

```python
import cv2
import numpy as np
from util import plot_images, normalize

image = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)  # Ensure the image is in grayscale

compass_kernels = [
    np.array([[1, 1, 1], [1, -2, 1], [-1, -1, -1]]),  # North
    np.array([[1, 1, 1], [-1, -2, 1], [-1, -1, 1]]),  # North-East
    np.array([[-1, 1, 1], [-1, -2, 1], [-1, 1, 1]]),  # East
    np.array([[-1, -1, 1], [-1, -2, 1], [1, 1, 1]]),  # South-East
    np.array([[-1, -1, -1], [1, -2, 1], [1, 1, 1]]),  # South
    np.array([[1, -1, -1], [1, -2, -1], [1, 1, 1]]),  # South-West
    np.array([[1, 1, -1], [1, -2, -1], [1, 1, -1]]), # West
    np.array([[1, 1, 1], [1, -2, -1], [1, -1, -1]])  # North-West 
]

# Convolve the image with each Compass operator kernel
compass_results = [cv2.filter2D(image, cv2.CV_64F, kernel) for kernel in compass_kernels]

# Calculate the gradient magnitude for each Compass direction and normalize the images
gradient_magnitudes = [np.abs(result) for result in compass_results]
gradient_magnitudes = [normalize(result) for result in gradient_magnitudes]

# Show the results with labels
max_image = np.max(gradient_magnitudes, axis=0)
# combine the images into a single image findind the max value for each pixel
results = [image] + gradient_magnitudes + [max_image]
result_names = ["Original Image", "North", "North-East", "East", "South-East", "South", "South-West", "West", "North-West", "Max"]
           

# Display the images in a grid
plot_images(results, result_names, 5, 2)
```
