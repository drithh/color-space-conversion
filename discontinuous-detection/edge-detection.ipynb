```python
import cv2
import numpy as np
from util import plot_images, normalize
%load_ext autoreload
%autoreload 2

# Membaca gambar
image = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)

# Tentukan operator Sobel
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

# Melakukan konvolusi gambar dengan filter Sobel
sobel_x_result = cv2.filter2D(image, cv2.CV_64F, sobel_x)
sobel_y_result = cv2.filter2D(image, cv2.CV_64F, sobel_y)

# Menghitung magnitudo gradien
gradient_magnitude = np.sqrt(sobel_x_result**2 + sobel_y_result**2)

# Menampilkan tepi vertikal yang dinormalisasi
result = [image, normalize(sobel_x_result), normalize(sobel_y_result), normalize(gradient_magnitude)]
result_name = ["Gambar Asli", "Sobel X", "Sobel Y", "Magnitudo Gradien"]

plot_images(result, result_name)
```

```python
import cv2
import numpy as np
from util import plot_images, normalize

# Membaca gambar
image = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)  # Pastikan gambar dalam skala abu-abu

# Definisikan kernel operator Compass
compass_kernels = [
    np.array([[1, 1, 1], [1, -2, 1], [-1, -1, -1]]),  # Utara
    np.array([[1, 1, 1], [-1, -2, 1], [-1, -1, 1]]),  # Timur Laut
    np.array([[-1, 1, 1], [-1, -2, 1], [-1, 1, 1]]),  # Timur
    np.array([[-1, -1, 1], [-1, -2, 1], [1, 1, 1]]),  # Tenggara
    np.array([[-1, -1, -1], [1, -2, 1], [1, 1, 1]]),  # Selatan
    np.array([[1, -1, -1], [1, -2, -1], [1, 1, 1]]),  # Barat Daya
    np.array([[1, 1, -1], [1, -2, -1], [1, 1, -1]]),  # Barat
    np.array([[1, 1, 1], [1, -2, -1], [1, -1, -1]])  # Barat Laut
]

# Melakukan konvolusi gambar dengan setiap kernel operator Compass
compass_results = [cv2.filter2D(image, cv2.CV_64F, kernel) for kernel in compass_kernels]

# Menghitung magnitudo gradien untuk setiap arah Compass dan menormalisasi gambar
gradient_magnitudes = [np.abs(result) for result in compass_results]
gradient_magnitudes = [normalize(result) for result in gradient_magnitudes]

# Mencari nilai maksimum untuk setiap piksel
max_image = np.max(gradient_magnitudes, axis=0)

# menggabungkan gambar-gambar menjadi satu gambar dengan mencari nilai maksimum untuk setiap piksel
results = [image] + gradient_magnitudes + [max_image]
result_names = ["Gambar Asli", "Utara", "Timur Laut", "Timur", "Tenggara", "Selatan", "Barat Daya", "Barat", "Barat Laut", "Maksimum"]

# Menampilkan gambar dalam bentuk grid
plot_images(results, result_names, 5, 2)
```
